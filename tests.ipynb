```python
from datasets.clevrer_ds import CLEVRERDataset
from torch.utils.data import DataLoader
from datasets.shapes_ds import generate_shape_dataset_torch
from datasets.calvin_dataset import CalvinDataset, CalvinDatasetSmall
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('generating random shapes dataset')
dataset = generate_shape_dataset_torch(num_images=5_000)
milestones = (20, 40, 80)
```

```python
dataloader = DataLoader(dataset, shuffle=True, batch_size=1, num_workers=0, pin_memory=True,drop_last=True)
pbar = tqdm(iterable=dataloader)
for batch in pbar:
    batch = batch[0]
    print(batch.shape)
    for i in range(batch.shape[0]):
        print(batch[i].shape)
        image = batch[i].permute(1, 2, 0).numpy()
        plt.imshow(image)
        plt.axis('off')
        plt.show()
        #print(image.shape)
    break
```
